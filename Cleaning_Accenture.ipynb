```python
import pandas as pd

# importing data
reactions = pd.read_csv('Reactions.csv')
reactiontypes = pd.read_csv('ReactionTypes.csv')
content = pd.read_csv('Content.csv')
```

```python
#checking data
print("Reactions DataFrame:")
print(reactions.head())

print("\nReaction Types DataFrame:")
print(reactiontypes.head())

print("\nContent DataFrame:")
print(content.head())
```

```python
#preparing cleanup of reactions df
print(reactions.columns)
print(reactions.dtypes)
print(reactions.isnull().sum())
```

```python
#remove first unnamed column
reactions.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
reactions.head(10)
```

```python
#convert to datetime
reactions['Datetime'] = pd.to_datetime(reactions['Datetime'])
#drop rows with missing type values
reactions.dropna(subset=['Type'], inplace=True)
```

```python
print(reactions.isnull().sum())
```

```python
#checking for duplicate rows
reactions = reactions.drop_duplicates()
print(reactions.isnull().sum())
```

```python
# display the cleaned df to verify changes
#changing Type column to lowercase
reactions['Type'] = reactions['Type'].str.lower()
print("Cleaned DataFrame:")
reactions.head()
cleanreactions = reactions
cleanreactions
```

```python
#prepping cleanup of content df
print(content.columns)
print(content.dtypes)
print(content.isnull().sum())
```

```python
content
```

```python
#drop unnamed index
content.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
#ensuring normativity within the columns table
content['Category'] = content['Category'].str.lower()
#ensuring that there is no quotation marks within the table
content['Category'] = content['Category'].str.replace('"', '', regex=False)
```

```python
#renaming null or nan columns as unknown  
content['URL'] = content['URL'].fillna('unknown')
print(content.isnull().sum())
#renaming the type column to content type 
content.rename(columns={'Type': 'Content Type'}, inplace=True)
```

```python
reactiontypes
```

```python
# dropping unnamed:0 index
reactiontypes.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
print(reactiontypes.isnull().sum())
```

```python

cleanreactiontypes = reactiontypes
cleancontent = content
cleanreactions.drop(columns = ['User ID'])


```

```python
cleanreactions.rename(columns={'Type': 'Reaction Type'}, inplace=True)
cleanreactiontypes.rename(columns={'Type': 'Reaction Type'}, inplace=True)
```

```python
#merging tables for export
merged_data = cleanreactions.merge(cleancontent, on='Content ID', how='left')


final_data = merged_data.merge(cleanreactiontypes, on='Reaction Type', how='left')

final_data.rename(columns={'Type': 'Content Type'}, inplace=True)
```

```python
#exporting dataset
final_data.to_csv('final_dataset.csv', index=False)
```

```python
#displaying the 5 top categories

category_scores = final_data.groupby('Category')['Score'].sum()

top_5_categories = category_scores.sort_values(ascending=False).head(5)

print(top_5_categories)
```

```python
# creating a dataframe for top 5 scores
top_5_categories_df = top_5_categories.reset_index().rename(columns={'score': 'total_score'})
```

```python
#creating excel file with top five categories and top 5 scores
with pd.ExcelWriter('final_dataset_with_top_5_categories.xlsx') as writer:
    final_data.to_excel(writer, sheet_name='Cleaned Dataset', index=False)
    top_5_categories_df.to_excel(writer, sheet_name='Top 5 Categories', index=False)
```
